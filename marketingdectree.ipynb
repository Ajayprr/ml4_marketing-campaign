###laterby we initiate our X & Y to train our model further

```python
ctree=tree.DecisionTreeClassifier()
ctree.fit(x_train,y_train)
```

will predict the model on x_test assign results to y_test_pred object

```python
##will use  gridsearchcv to find best suiting parameters for our model
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={"criterion":['gini','entropy'],
            "max_features":[1,2,3,4,5,6],
            "splitter":['best','random'],
            "min_samples_leaf":[1,2,3,4,5,6],
            "max_depth":[1,2,3,4]
           
           }
```

```python
opt=GridSearchCV(ctree,param_grid,scoring='accuracy',verbose=2,cv=5)
```

```python
opt.fit(x_train,y_train)
```

```python
opt.best_params_
```

```python
opt.best_score_
```

```python
finalmodel=opt.best_estimator_
```

```python
y_test_predmodel=finalmodel.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_predmodel)
```

```python
metrics.recall_score(y_test_predmodel,y_test)
```

```python
metrics.precision_score(y_test_predmodel,y_test)
```
